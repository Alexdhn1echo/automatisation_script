```python
from pynput import mouse, keyboard

# Initialiser le compteur
click_count = 0

def on_click(x, y, button, pressed):
    global click_count  # Utiliser la variable globale
    if pressed:
        click_count += 1  # IncrÃ©menter le compteur Ã  chaque clic
        if button == mouse.Button.left:
            print(f"Clic gauche Ã  la position: ({x}, {y}), Compteur: {click_count}")
        elif button == mouse.Button.right:
            print(f"Clic droit Ã  la position: ({x}, {y}), Compteur: {click_count}")

def on_press(key):
    if key == keyboard.Key.space:
        mouse_listener.stop()
        keyboard_listener.stop()

mouse_listener = mouse.Listener(on_click=on_click)
keyboard_listener = keyboard.Listener(on_press=on_press)

mouse_listener.start()
keyboard_listener.start()  

mouse_listener.join()
keyboard_listener.join()
 
```
